```python
import requests
from bs4 import BeautifulSoup
import csv

title = []
location = []
company = []
jskills = []

result = requests.get("https://wuzzuf.net/search/jobs/?q=python&a=hpb")
# save content
src = result.content
#print(soup.prettify)
soup = BeautifulSoup(src, 'lxml')


job_title = soup.find_all('h2', {'class': 'css-m604qf'})
#print(job_title)
company_title = soup.find_all("a", {'class': 'css-17s97q8'})
#print(company_title)
locations = soup.find_all('span', {'class': 'css-5wys0k'})
#print(locations)
skills = soup.find_all('div', {'class': 'css-y4udm8'})
#print(skills)

for i in range(len(job_title)):
    title.append(job_title[i].text.strip())  # Strip whitespace
    location.append(locations[i].text.strip())  # Strip whitespace
    company.append(company_title[i].text.strip())  # Strip whitespace
    jskills.append(skills[i].text.strip())  # Strip whitespace

file_list = [title, company, location, jskills]

with open("C:\Users\clt\Desktop\Web Scraping\1-2.csv", 'w', encoding="latin1", newline='') as my_file:  # Add 'newline' parameter
    wr = csv.writer(my_file)
    wr.writerow(['Job Title', 'Company Name', 'Location', 'Skills'])
    wr.writerows(zip(title, company, location, jskills))
```
