# Line Plot

```python
#one-variable line chart
# importing the required libraries 
import matplotlib.pyplot as plt 
import numpy as np 
  
# define data values 
x = np.array([1,6,4,7,14,18,21])  # X-axis points 
  
plt.plot(x)  # Plot the chart
plt.show()  # display
```

```python
#two-variable line chart
# importing the required libraries 
import matplotlib.pyplot as plt 
import numpy as np 
  
# define data values 
x = np.array([1,3,6,10,17,18,21])  # X-axis points 
y = np.array([3,5,1,7,9,2,4])  # Y-axis points 
  
plt.plot(x,y)  # Plot the chart
plt.show()  # display
```

```python
#two-variable line chart
#example2
x = [1,2,4,5,6]
y = [2,4,8,10,12]

plt.plot(x,y, label = "2x" , color = "black", linewidth = 2 , marker = "*", linestyle=":")
plt.title("this is a good graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
  
x = np.array([1, 2, 3, 4]) 
y = x*2
  
# first plot with X and Y data 
plt.plot(x, y) 
  
x1 = [2, 4, 6, 8] 
y1 = [3, 5, 7, 9] 

# second plot with x1 and y1 data 
plt.plot(x1, y1, '-.') #fmt

#we can name x and y
plt.xlabel("X-axis data") 
plt.ylabel("Y-axis data")

#we can add title to our plot
plt.title('multiple plots') 

plt.show()
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,40]
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend()
plt.show()
```

# Bar Chart

```python
import numpy as np
import matplotlib.pyplot as plt 


# creating the dataset
data = {'C':20, 'C++':15, 'Java':30,'Python':35}
courses = list(data.keys())
values = list(data.values())

plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(courses, values, color ='maroon', width = 0.4)

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
lables = ["A", "B", "C"]
values = [1,4 ,2]
bars = plt.bar(lables, values, color = "Gray")
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C']
values1 = [3, 5, 2]
values2 = [2, 4, 1]

plt.bar(categories, values1, label='Group 1', color='blue', alpha=0.7)
plt.bar(categories, values2, label='Group 2', color='orange', alpha=0.7, bottom=values1)

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Plot')
plt.legend()

plt.show()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,85,27]
profit=[40,7,34,12]
xpos = np.arange(len(company)) # numpyarray : 0 , 1 , 2, 3

plt.bar(company,revenue, width=0.4, label="Revenue") # x = 0, x = 0.4 
plt.bar(xpos-0.4,profit, width=0.4,label="Profit")# x = -.4  , x = 0
# 0 - 0.4 = -.4 , .6 , 1.6 , 2.6 
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

```python
#3D Scatter
plt.rcParams['figure.figsize'] = (8,6)
np.random.seed(31)
mu = 3
n=50

x = np.random.normal(mu, 1, size=n)
y = np.random.normal(mu, 1, size=n)
z = np.random.normal(mu, 1, size=n)
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(45, 215)
```

# Box Plot

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np


# Creating dataset
np.random.seed(10)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize =(10, 7))

# Creating plot
plt.boxplot(data)

# show plot
plt.show()
```

```python
# Creating dataset
np.random.seed(10)
 
data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]
 
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data)
 
# show plot
plt.show()
```

# Violin Plot

```python
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)

## combine these different collections into a list
data_to_plot = [collectn_1, collectn_2, collectn_3, collectn_4]

# Create a figure instance
fig = plt.figure()

# Create an axes instance
ax = fig.add_axes([0,0,1,1])

# Create the boxplot
bp = ax.violinplot(data_to_plot)
plt.show()
```

```python
pokemon=pd.read_csv("Pokemon.csv")
```

```python
# bulbasaur = pokemon[['Name', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']][pokemon.loc[:, 'Name'] == 'Bulbasaur']
poke_num = pokemon[['Name', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].set_index('Name')

bulbasaur = poke_num.loc['Bulbasaur']
charmander = poke_num.loc['Charmander']
squirtle = poke_num.loc['Squirtle']
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
fig.suptitle('Initial Pokemon - 1st Generation')

# Bulbasaur
sns.barplot(ax=axes[0], x=bulbasaur.index, y=bulbasaur.values)
axes[0].set_title(bulbasaur.name)

# Charmander
sns.barplot(ax=axes[1], x=charmander.index, y=charmander.values)
axes[1].set_title(charmander.name)

# Squirtle
sns.barplot(ax=axes[2], x=squirtle.index, y=squirtle.values)
axes[2].set_title(squirtle.name)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

fig.suptitle('Pokemon Stats by Generation')

sns.boxplot(ax=axes[0, 0], data=pokemon, x='Generation', y='Attack')
sns.boxplot(ax=axes[0, 1], data=pokemon, x='Generation', y='Defense')
sns.boxplot(ax=axes[0, 2], data=pokemon, x='Generation', y='Speed')
sns.boxplot(ax=axes[1, 0], data=pokemon, x='Generation', y='Sp. Atk')
sns.boxplot(ax=axes[1, 1], data=pokemon, x='Generation', y='Sp. Def')
sns.boxplot(ax=axes[1, 2], data=pokemon, x='Generation', y='HP')
```
